```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import random
#pip install networkx==2.4
```

```python
g=nx.erdos_renyi_graph(10, 0.3, seed=None, directed=False)
plt.figure(1,figsize=(5,5)) 
pos=nx.kamada_kawai_layout(g)
nx.draw(g,pos,with_labels=True)
plt.show()
```

```python
for i in g.nodes():
    temp=random.random()
    if temp<=0.5:
        g.nodes[i]['gender'] = 'female'
    else:
        g.nodes[i]['gender'] = 'male'
```

```python
g.nodes(data='gender')
```

```python
colors=[]
for i in g.nodes():
    if g.nodes[i]['gender'] == 'female':
        colors.append('r')
    else:
        colors.append('g')
```

```python
plt.figure(1,figsize=(5,5)) 
pos=nx.kamada_kawai_layout(g)
nx.draw(g,pos,with_labels=True,node_color=colors)
plt.show()
```
