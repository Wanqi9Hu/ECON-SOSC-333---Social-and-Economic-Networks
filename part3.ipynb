```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import math 
#pip install networkx==2.4
```

```python
def char(g):
    node_list=nx.nodes(g)
    clustering=[nx.clustering(g,n) for n in node_list] 
    degree=[g.degree(n) for n in node_list] 
    weighted_degree=[g.degree(n, weight='weight') for n in node_list] 
    closeness=[nx.closeness_centrality(g,n) for n in node_list]
    eigenvc=[]
    t=nx.eigenvector_centrality(g)
    for k in t.keys():
        eigenvc.append(t[k])
    df=pd.DataFrame({'nodes': node_list, 'degree': degree, 'weighted degree': weighted_degree,'closeness': closeness, 'clustering': clustering,'eigenvector': eigenvc })    
    return df
```

# Erdos-Renyi random graph

```python
g=nx.erdos_renyi_graph(50, 0.2, seed=None, directed=False)
plt.figure(1,figsize=(5,5)) 
pos=nx.kamada_kawai_layout(g)
nx.draw(g,pos,with_labels=True)
plt.show()
```

```python
char(g).head(10)
```
